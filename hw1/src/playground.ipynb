```python
log = open('./data/log/robotdata1.log', 'r')
log_data = log.readlines()
print(len(log_data))
```

```python
sensor_readings = []
for l in log_data:
    l = l.strip()
    tokens = l.split(' ')
    data = tokens[7:-1]
    if tokens[0] == 'L':
         sensor_readings.extend(data)

sensor_readings = np.array(sensor_readings, 'float')
print(sensor_readings.max(), sensor_readings.min())
plt.hist((sensor_readings).flatten())
```

```python
import multiprocessing

print("No of cpus = ", multiprocessing.cpu_count())
```

```python
from multiprocessing import Process

def print_func(continent = 'Asia'):
    print('Name of continent is : ', continent)
    
names = ['North America', 'India', 'Europe']
procs = []
proc = Process(target=print_func)
procs.append(proc)
proc.start()

for name in names:
    proc = Process(target=print_func, args=(name,))
    procs.append(proc)
    proc.start()
    
for proc in procs:
    proc.join()
```
